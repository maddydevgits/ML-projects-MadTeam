```python
import numpy as np
import matplotlib.pyplot as plt
import os
import warnings
from sklearn.cluster import KMeans

os.environ['OMP_NUM_THREADS'] = '1'
warnings.filterwarnings("ignore", category=UserWarning)

np.random.seed(123)

# Function to calculate total intra-cluster sum of squares
def iss(k):
    kmeans = KMeans(n_clusters=k, max_iter=100, n_init=100, algorithm="auto")
    kmeans.fit(customer_data.iloc[:, 2:5])
    return kmeans.inertia_

k_values = np.arange(1, 11)
iss_values = np.vectorize(iss)(k_values)

plt.plot(k_values, iss_values, 'bo-', markerfacecolor='none')
plt.xlabel("Number of clusters K")
plt.ylabel("Total intra-cluster sum of squares")
plt.title("Elbow Method for Optimal K")
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
import matplotlib.pyplot as plt
import numpy as np

k = 2
kmeans = KMeans(n_clusters=k, max_iter=100, n_init=50, algorithm="auto")
kmeans.fit(customer_data.iloc[:, 2:5])

cluster_labels = kmeans.predict(customer_data.iloc[:, 2:5])
silhouette_avg = silhouette_score(customer_data.iloc[:, 2:5], cluster_labels)

sample_silhouette_values = silhouette_samples(customer_data.iloc[:, 2:5], cluster_labels)

fig, ax = plt.subplots(figsize=(6, 4))

y_lower = 10
for i in range(k):
    ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]
    ith_cluster_silhouette_values.sort()

    size_cluster_i = ith_cluster_silhouette_values.shape[0]
    y_upper = y_lower + size_cluster_i

    ax.fill_betweenx(np.arange(y_lower, y_upper), 0, ith_cluster_silhouette_values, alpha=0.7)
    ax.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
    y_lower = y_upper + 10

ax.set_title("Silhouette plot")
ax.set_xlabel("Silhouette coefficient values")
ax.set_ylabel("Cluster label")

ax.axvline(x=silhouette_avg, color="red", linestyle="--")

plt.show()
```
